```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Laden Datensatzes
taxi_data = sns.load_dataset("taxis")
taxi_data.to_csv("taxis.csv")
taxi_data.head()
```

```python
# Erstellen des Boxplots f√ºr "total"
plt.figure(figsize=(8, 6))
sns.boxplot(x="total", data=taxi_data)
plt.title("Boxplot von Total")
plt.xlabel("Total")
plt.show()
```
