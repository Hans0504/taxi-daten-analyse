```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Laden Datensatzes
taxi_data = sns.load_dataset("tips")
print(taxi_data.head())

# Speichern des Datensatzes in einer CSV-Datei
#taxi_data.to_csv("taxi_data.csv", index=False)
```

```python
# Erstellen der neuen Spalte "total" mit Addition aus "total-bill" und "tip"
taxi_data["total"] = taxi_data["total_bill"] + taxi_data["tip"]
taxi_data
```

```python
# Erstellen des Boxplots fÃ¼r "total"
plt.figure(figsize=(8, 6))
sns.boxplot(x="total", data=taxi_data)
plt.title("Boxplot von Total")
plt.xlabel("Total")
plt.show()
```
